# PREPROCESSING

## ENCODING

```python
from sklearn.preprocessing import LabelEncoder

# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Menggunakan label encoder untuk 'Area_code'
data['area_code'] = label_encoder.fit_transform(data['area_code'])
```

```python
# Membuat dictionary mapping dari setiap nilai unik 'State' ke angka
state_encoding = {state: i for i, state in enumerate(data['state'].unique())}

# Menggunakan dictionary mapping untuk mengganti nilai dalam kolom 'State'
data['state'] = data['state'].replace(state_encoding)
```

```python
# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Menggunakan label encoder untuk 'International_plan'
data['international_plan'] = label_encoder.fit_transform(data['international_plan'])

# Menggunakan label encoder untuk 'Voice_mail_plan'
data['voice_mail_plan'] = label_encoder.fit_transform(data['voice_mail_plan'])

# Menggunakan label encoder untuk 'Churn'
data['churn'] = label_encoder.fit_transform(data['churn'])
```

```python
data.head()
```

Outliner ICR

```python
from scipy import stats

print(f'Jumlah baris sebelum memfilter outlier: {len(data)}')

# Membuat array boolean dengan ukuran yang sama dengan jumlah baris di DataFrame
filtered_entries = np.array([True] * len(data))

# Memilih semua kolom dalam DataFrame
all_cols = data.columns

# Iterasi melalui setiap kolom untuk memfilter outlier menggunakan z-score
for col in all_cols:
    zscore = np.abs(stats.zscore(data[col]))
    filtered_entries = (zscore < 3) & filtered_entries

# Memfilter DataFrame dengan menggunakan array boolean yang sudah dibuat
data = data[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(data)}')
```

### Preprocessing selesai

MODEL

```python
x= data.loc[:, data.columns != 'churn']
```

```python
Y = data['churn']
```

```python
from sklearn.model_selection import train_test_split

# split data into train and test


X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size=0.4, random_state=200)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Menampilkan jumlah kelas sebelum oversampling
counter = Counter(y_train)
print('Before:', counter)

# Inisialisasi SMOTE
smt = SMOTE(random_state=42)

# Melakukan oversampling pada data training
x_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)

# Menampilkan jumlah kelas setelah oversampling
counter = Counter(y_train_sm)
print('After:', counter)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=200)
clf.fit(X_train, y_train)
y_pred2 = clf.predict(X_test)
print('Accuracy score:', accuracy_score(y_test, y_pred2) * 100)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred2))
print('Classification Report:\n', classification_report(y_test, y_pred2))
```

Mengetahui Nilai ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

models = [
    {
        'name': 'Logistic Regression',
        'model': LogisticRegression(random_state=200)
    },
    {
        'name': 'Random Forest',
        'model': RandomForestClassifier(random_state=200)
    },
    {
        'name': 'Decision Tree (Gini)',
        'model': DecisionTreeClassifier(criterion="gini", random_state=200, max_depth=3, min_samples_leaf=5)
    }
]


plt.figure(figsize=(8, 6))
for model in models:

    model['model'].fit(X_train, y_train)


    y_pred_proba = model['model'].predict_proba(X_test)[:, 1]


    roc_auc = roc_auc_score(y_test, y_pred_proba)


    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)


    plt.plot(fpr, tpr, label=f'{model["name"]} (ROC AUC = {roc_auc:.2f})')


plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

### Testing

```python
import pandas as pd
import joblib
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Load trained model
model = joblib.load("random-forest_churn.pkl")

# Define categorical and numeric features
categorical_features = ['state', 'area_code', 'international_plan', 'voice_mail_plan']
numeric_features = ['account_length', 'number_vmail_messages', 'total_day_minutes', 'total_day_calls',
                    'total_day_charge', 'total_eve_minutes', 'total_eve_calls', 'total_eve_charge',
                    'total_night_minutes', 'total_night_calls', 'total_night_charge', 'total_intl_minutes',
                    'total_intl_calls', 'total_intl_charge', 'number_customer_service_calls']

# Define transformer for categorical features
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('ordinal', OrdinalEncoder())
])

# Define transformer for numeric features
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

# Combine all transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num', numeric_transformer, numeric_features)
    ])

from google.colab import files
uploaded = files.upload()

import io
data_to_predict = pd.read_csv(io.BytesIO(uploaded['Data Test.csv']), sep=";", encoding='latin1')

# Fit and transform preprocessing on the dataset to be predicted
X_to_predict = preprocessor.fit_transform(data_to_predict)

# Perform churn prediction on preprocessed data
predicted_churn = model.predict(X_to_predict)

# Add predicted churn column to dataset
data_to_predict['predicted_churn'] = predicted_churn

# Show or save modified dataset
print(data_to_predict)  # Show modified dataset
data_to_predict.to_csv("hasil_prediksi_churn.csv", index=False)  # Save modified dataset to CSV file
```
